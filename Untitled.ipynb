```python
import pandas as pd

results = pd.read_csv('results.csv')
results.head()
```

```python
results.describe()
```

```python
counts = results.groupby('country')
counts = counts.apply(lambda df: df['date'].count())
```

```python
max_matches_country = counts[counts==counts.max()].index[0]
print(max_matches_country)
```

```python
counts[counts==counts.min()]
```

```python
counts[counts>counts.mean()]
```
